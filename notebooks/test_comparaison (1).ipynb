```python
import json
import urllib.request
import time
from IPython.display import display, Markdown

OLLAMA_URL   = "http://localhost:11434/api/generate"
OLLAMA_MODEL = "mistral"

ZONES_TEST = {
    "Méditerranée": "Mer Méditerranée, mer Égée, mer Adriatique, détroit de Gibraltar, archipel grec",
    "Océan Pacifique Sud": "Océan Pacifique Sud, Polynésie, Mélanésie, côtes australiennes"
}

def appeler_ollama(prompt, num_predict=2000):
    payload = json.dumps({
        "model": OLLAMA_MODEL,
        "prompt": prompt,
        "stream": False,
        "options": {"temperature": 0.2, "num_predict": num_predict}
    }).encode("utf-8")
    req = urllib.request.Request(
        OLLAMA_URL, data=payload,
        headers={"Content-Type": "application/json"}, method="POST"
    )
    try:
        with urllib.request.urlopen(req, timeout=180) as resp:
            return json.loads(resp.read().decode("utf-8")).get("response", "").strip()
    except Exception as e:
        return f"Erreur Ollama : {e}"

def prompt_comparaison(nom_zone, description_zone):
    return f"""Tu es un scientifique marin et historien spécialisé dans l'environnement océanique.

ZONE GÉOGRAPHIQUE : {nom_zone}
DESCRIPTION : {description_zone}

CONTEXTE IMPORTANT : En 1866, Jules Verne publiait "Vingt mille lieues sous les mers". À cette époque, l'exploration sous-marine en était à ses balbutiements — les premiers sous-marins rudimentaires comme le Plongeur (1863) venaient d'apparaître, mais aucun n'était capable d'une navigation autonome prolongée. Jules Verne imaginait avec une vision extraordinairement prophétique un sous-marin électrique capable de parcourir tous les océans du globe. Il faut souligner ce génie visionnaire.

TÂCHE : Rédige une comparaison entre la connaissance maritime de cette zone en 1866 et aujourd'hui.

STRUCTURE OBLIGATOIRE en 4 sections :

⚓ AU XIXe SIÈCLE (vers 1866)
Décris les techniques de navigation, les outils disponibles, les connaissances scientifiques et l'état de la faune marine à cette époque. Mentionne brièvement le génie visionnaire de Jules Verne qui imaginait explorer ces eaux à bord d'un sous-marin électrique alors que la technologie n'existait pas encore.

🌊 AUJOURD'HUI
Décris l'état actuel de l'écosystème marin, les nouvelles espèces découvertes, les technologies modernes d'exploration sous-marine (ROV, bathyscaphes, satellites...).

⚠️ PROBLÈMES ENVIRONNEMENTAUX ACTUELS
Cite les problèmes spécifiques à cette zone : surpêche, réchauffement climatique, acidification, pollution plastique, espèces menacées.

📜 LOIS ET PROTECTIONS EN VIGUEUR
Cite les lois de pêche, accords internationaux et zones marines protégées actuellement en vigueur dans cette zone.

RÈGLES :
- Faits réels et vérifiables uniquement
- Texte littéraire, fluide et élégant — pas de tableaux ni de listes à puces
- Chaque section fait 3 à 5 phrases
- Réponse en français uniquement
- Jules Verne peut être mentionné dans la section XIXe siècle uniquement
"""

resultats = {}

for nom_zone, description in ZONES_TEST.items():
    print(f"\n{'='*55}")
    print(f"⚖️  {nom_zone}")
    print(f"{'='*55}")
    print("🤖 Ollama en cours...")

    reponse = appeler_ollama(prompt_comparaison(nom_zone, description))
    resultats[nom_zone] = reponse

    display(Markdown(f"""
---
### ⚖️ {nom_zone}
{reponse}
---
"""))
    time.sleep(2)

print("\n✅ Test terminé !") 
```
