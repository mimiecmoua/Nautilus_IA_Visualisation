```python
import json
import urllib.request
import time
from pathlib import Path
from IPython.display import display, Markdown

# =========================================================
# CONFIG
# =========================================================
OLLAMA_URL   = "http://localhost:11434/api/generate"
OLLAMA_MODEL = "mistral"

# ── 2 zones de test ──
ZONES_TEST = {
    "Méditerranée": {
        "description": "Mer Méditerranée, mer Égée, mer Adriatique, détroit de Gibraltar",
        "escales": ["L'Archipel grec", "Méditerranée", "L'Atlantide", "Isthme de Suez"],
        "couleur": "#b8e0d4"
    },
    "Océan Pacifique Sud": {
        "description": "Océan Pacifique Sud, archipels de Polynésie, Mélanésie, Micronésie, côtes australiennes",
        "escales": ["Île de Crespo", "Archipel des Pomotou", "Archipel des nouvelles Hébrides", "La Pouasie", "Cap Wessel", "Mer de Timor"],
        "couleur": "#b8d4e0"
    }
}

# =========================================================
# OLLAMA
# =========================================================
def appeler_ollama(prompt, num_predict=2000):
    payload = json.dumps({
        "model": OLLAMA_MODEL,
        "prompt": prompt,
        "stream": False,
        "options": {"temperature": 0.2, "num_predict": num_predict}
    }).encode("utf-8")
    req = urllib.request.Request(
        OLLAMA_URL, data=payload,
        headers={"Content-Type": "application/json"}, method="POST"
    )
    try:
        with urllib.request.urlopen(req, timeout=180) as resp:
            return json.loads(resp.read().decode("utf-8")).get("response", "").strip()
    except Exception as e:
        return f"Erreur Ollama : {e}"

def prompt_decouvertes(nom_zone, description_zone):
    return f"""Tu es un historien des sciences spécialisé dans le XIXe siècle et l'exploration maritime.

ZONE GÉOGRAPHIQUE : {nom_zone}
DESCRIPTION : {description_zone}

TÂCHE : Liste les grandes découvertes scientifiques maritimes du XIXe siècle (1800-1900) liées à cette zone géographique.

DOMAINES À COUVRIR :
- Naturalistes et biologistes ayant étudié cette zone (Darwin, Milne-Edwards, Haeckel, Agassiz...)
- Expéditions océanographiques réelles dans cette zone
- Espèces maritimes découvertes et décrites pour la première fois
- Avancées en navigation et cartographie maritime
- Découvertes géologiques et bathymétriques sous-marines

RÈGLES STRICTES :
- Uniquement des faits historiques réels et vérifiables du XIXe siècle
- Cite toujours le nom du scientifique, la date et la découverte précise
- Réponse en français uniquement
- Ne mentionne pas Jules Verne ni son roman

FORMAT — pour chaque découverte :
🔬 [Nom du scientifique] ([année]) — [découverte précise]
   Contexte : [une phrase de contexte]
"""

# =========================================================
# TEST SUR 2 ZONES
# =========================================================
resultats = {}

for nom_zone, info in ZONES_TEST.items():
    print(f"\n{'='*55}")
    print(f"🌊 Zone : {nom_zone}")
    print(f"📍 Escales : {', '.join(info['escales'])}")
    print(f"{'='*55}")
    print("🤖 Envoi à Ollama...")

    reponse = appeler_ollama(prompt_decouvertes(nom_zone, info["description"]))
    resultats[nom_zone] = reponse

    display(Markdown(f"""
---
### 🌊 {nom_zone}
{reponse}
---
"""))
    time.sleep(2)

print("\n✅ Test terminé !")
```
