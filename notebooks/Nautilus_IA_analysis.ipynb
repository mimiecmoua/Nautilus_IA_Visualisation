```python
# Importation des bibliothèques
import pandas as pd
from sklearn.cluster import KMeans

# 1. Charger les données
df = pd.read_csv("../data/Relever-Nautilus-Position-UTF84.csv", delimiter=";", encoding="utf-8")

# Afficher les premières lignes pour vérifier
print("Données chargées :")
print(df.head())

# 2. Appliquer K-Means (clustering)
# On utilise uniquement les colonnes Latitude_Décimal et Longitude_Décimal pour le clustering
kmeans = KMeans(n_clusters=3)  # On choisit 3 clusters (tu peux ajuster ce nombre)
df['Cluster'] = kmeans.fit_predict(df[['Latitude_Décimal', 'Longitude_Décimal']])

# Afficher les clusters attribués
print("\nClusters attribués :")
print(df[['Escales du Nautilus', 'Event', 'Cluster']])

# 3. Exporter les résultats en JSON
df.to_json('escales_clusters.json', orient='records', force_ascii=False)

print("\nRésultats exportés dans 'escales_clusters.json'")
```
