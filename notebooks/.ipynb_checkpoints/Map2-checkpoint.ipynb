```python
import folium
from folium.plugins import MarkerCluster
import json
import pandas as pd
from math import radians, cos, sin, sqrt

# Chargement des données
with open('../data/escales.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)

# Fonction de calcul de distance au centre de la Terre
def distance_from_center(lat):
    R = 6371e3  # rayon moyen en mètres
    lat_rad = radians(lat)
    return sqrt((R * cos(lat_rad))**2 + (R * sin(lat_rad))**2)

# Ajout des distances
df['Distance_Centre_Terre'] = df['Latitude_Décimal'].apply(distance_from_center).astype(int)

# Initialisation de la carte
m = folium.Map(location=[0, 0], zoom_start=2, tiles='CartoDB positron')
cluster_layer = MarkerCluster().add_to(m)

coordinates = []
for _, row in df.iterrows():
    coord = [row['Latitude_Décimal'], row['Longitude_Décimal']]
    coordinates.append(coord)
    
    popup_text = (
        f"<b>{row['Escales du Nautilus']}</b><br>"
        f"📅 Date : {row['Date']}<br>"
        f"🌊 Océan/Mer : {row['Océan/Mer']}<br>"
        f"📖 Événement : {row['Event']}<br>"
        f"🧭📌 Latitude : {row['Latitude_Décimal']}<br>"
        f"🧭📌 Longitude : {row['Longitude_Décimal']}<br>"
        f"🌍🧠 Distance du centre de la Terre : {row['Distance_Centre_Terre']:,} mètres"
    )
    
    folium.Marker(
        location=coord,
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color='blue')
    ).add_to(cluster_layer)

# Tracé de la route
folium.PolyLine(locations=coordinates, color='blue', weight=3, opacity=0.7).add_to(m)

# Ajouter des icônes au début et à la fin du trajet
start_coord = coordinates[0]  # Première escale (début)
end_coord = coordinates[-1]   # Dernière escale (fin)

# Icône de début
folium.Marker(
    location=start_coord,
    popup="Début du voyage",
    icon=folium.DivIcon(
        icon_size=(30, 30),  # Taille de l'icône
        icon_anchor=(15, 15),  # Position de l'icône
        html='<div style="font-size: 12px; color: green; font-weight: bold;">Début</div>'  # Contenu de l'icône
    )
).add_to(m)

# Icône de fin
folium.Marker(
    location=end_coord,
    popup="Fin du voyage",
    icon=folium.DivIcon(
        icon_size=(30, 30),  # Taille de l'icône
        icon_anchor=(15, 15),  # Position de l'icône
        html='<div style="font-size: 12px; color: red; font-weight: bold;">Fin</div>'  # Contenu de l'icône
    )
).add_to(m)

# Sauvegarde de la carte
m.save("carte_nautilus_complete.html")
print("✅ Carte avec popups + icônes début/fin générée dans 'carte_nautilus_complete.html'")








```
