```python
import json
import urllib.request
import time
from IPython.display import display, Markdown

OLLAMA_URL   = "http://localhost:11434/api/generate"
OLLAMA_MODEL = "mistral"

# ── 9 zones océaniques ──
ZONES = {
    "Océan Pacifique Sud": "Océan Pacifique Sud, Polynésie, Mélanésie, Micronésie, côtes australiennes, îles Galápagos",
    "Océan Indien": "Océan Indien, côtes indiennes, Sri Lanka, île Keeling, golfe du Bengale, archipel des Maldives",
    "Mer Rouge": "Mer Rouge, golfe d'Aden, détroit de Babel-Mandeb, côtes arabiques et africaines orientales",
    "Méditerranée": "Mer Méditerranée, mer Égée, mer Adriatique, détroit de Gibraltar, archipel grec",
    "Océan Atlantique Sud": "Océan Atlantique Sud, mer des Sargasses, côtes africaines, côtes sud-américaines, mer des Caraïbes, Antilles",
    "Pôle Sud": "Océan Antarctique, mer de Weddell, banquise antarctique, terres australes",
    "Océan Atlantique Nord": "Océan Atlantique Nord, banc de Terre-Neuve, Gulf Stream, côtes américaines",
    "Mers Celtiques et Manche": "Manche, mer Celtique, côtes anglaises, côtes françaises, mer d'Irlande",
    "Mer de Norvège et Maelström": "Mer de Norvège, mer du Nord, archipel des Lofoten, Maelström, côtes scandinaves"
}

def appeler_ollama(prompt, num_predict=2000):
    payload = json.dumps({
        "model": OLLAMA_MODEL,
        "prompt": prompt,
        "stream": False,
        "options": {"temperature": 0.2, "num_predict": num_predict}
    }).encode("utf-8")
    req = urllib.request.Request(
        OLLAMA_URL, data=payload,
        headers={"Content-Type": "application/json"}, method="POST"
    )
    try:
        with urllib.request.urlopen(req, timeout=180) as resp:
            return json.loads(resp.read().decode("utf-8")).get("response", "").strip()
    except Exception as e:
        return f"Erreur Ollama : {e}"

def prompt_comparaison(nom_zone, description_zone):
    return f"""Tu es un scientifique marin et historien spécialisé dans l'environnement océanique.

ZONE GÉOGRAPHIQUE : {nom_zone}
DESCRIPTION : {description_zone}

CONTEXTE IMPORTANT : En 1866, Jules Verne publiait "Vingt mille lieues sous les mers". À cette époque, l'exploration sous-marine en était à ses balbutiements — les premiers sous-marins rudimentaires comme le Plongeur (1863) venaient d'apparaître, mais aucun n'était capable d'une navigation autonome prolongée. Jules Verne imaginait avec une vision extraordinairement prophétique un sous-marin électrique capable de parcourir tous les océans du globe. Il faut souligner ce génie visionnaire.

TÂCHE : Rédige une comparaison entre la connaissance maritime de cette zone en 1866 et aujourd'hui.

STRUCTURE OBLIGATOIRE en 4 sections :

⚓ AU XIXe SIÈCLE (vers 1866)
Décris les techniques de navigation, les outils disponibles, les connaissances scientifiques et l'état de la faune marine à cette époque dans cette zone spécifique. Mentionne brièvement le génie visionnaire de Jules Verne qui imaginait explorer ces eaux à bord d'un sous-marin électrique alors que la technologie n'existait pas encore.

🌊 AUJOURD'HUI
Décris l'état actuel de l'écosystème marin de cette zone, les nouvelles espèces découvertes, les technologies modernes d'exploration sous-marine (ROV, bathyscaphes, satellites...).

⚠️ PROBLÈMES ENVIRONNEMENTAUX ACTUELS
Cite les problèmes SPÉCIFIQUES à cette zone : espèces menacées propres à cette région, impacts particuliers du réchauffement, pollutions locales.

📜 LOIS ET PROTECTIONS EN VIGUEUR
Cite les lois de pêche, accords internationaux et zones marines protégées SPÉCIFIQUES à cette zone géographique.

RÈGLES :
- Faits réels et vérifiables uniquement
- Texte littéraire, fluide et élégant — pas de tableaux ni de listes à puces
- Chaque section fait 3 à 5 phrases
- Réponse en français uniquement
- Jules Verne mentionné dans la section XIXe siècle uniquement
- Chaque zone doit avoir des problèmes et lois SPÉCIFIQUES à sa géographie — pas de réponses génériques
"""

# ── Génération des 9 zones ──
resultats = {}

for nom_zone, description in ZONES.items():
    print(f"\n{'='*55}")
    print(f"⚖️  {nom_zone}")
    print(f"{'='*55}")
    print("🤖 Ollama en cours...")

    reponse = appeler_ollama(prompt_comparaison(nom_zone, description))
    resultats[nom_zone] = reponse

    display(Markdown(f"""
---
### ⚖️ {nom_zone}
{reponse}
---
"""))
    time.sleep(2)

# ── Sauvegarde JSON ──
with open("comparaison_zones.json", "w", encoding="utf-8") as f:
    json.dump(resultats, f, ensure_ascii=False, indent=2)

print("\n✅ Toutes les zones générées !")
print("📄 Sauvegardé dans comparaison_zones.json")
```
