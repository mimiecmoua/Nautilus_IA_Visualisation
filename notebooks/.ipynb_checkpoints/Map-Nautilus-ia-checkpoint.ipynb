```python
import json
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from sklearn.cluster import KMeans
from collections import Counter
import nltk

# 📌 Télécharger les ressources nécessaires pour l'analyse NLP
nltk.download('punkt', download_dir='C:/Users/Utilisateur/AppData/Roaming/nltk_data')
nltk.download('stopwords', download_dir='C:/Users/Utilisateur/AppData/Roaming/nltk_data')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# 📌 Étape 1 : Chargement des données JSON
with open('../data/escales.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

df = pd.DataFrame(data)

# 📌 Étape 2 : Analyse sémantique des événements
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('french'))

def extract_keywords(event):
    words = word_tokenize(event.lower())
    words = [word for word in words if word.isalpha() and word not in stop_words]
    return Counter(words).most_common(3)  # Top 3 mots clés

df['Mots_Clés'] = df['Event'].apply(lambda x: extract_keywords(str(x)))

# 📌 Étape 3 : Clustering géographique des escales
coords = df[['Latitude_Décimal', 'Longitude_Décimal']]
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10).fit(coords)
df['Cluster'] = kmeans.labels_

# 📌 Étape 4 : Création de la carte interactive
m = folium.Map(location=[0, 0], zoom_start=2, tiles='CartoDB positron')
cluster_layer = MarkerCluster().add_to(m)

# 📌 Étape 5 : Ajout des escales avec pop-ups enrichis
for _, row in df.iterrows():
    popup_text = (
        f"<b>{row['Escales du Nautilus']}</b><br>"
        f"📅 Date : {row['Date']}<br>"
        f"🌊 Océan/Mer : {row['Océan/Mer']}<br>"
        f"📖 Événement : {row['Event']}<br>"
        f"🧠 Mots-clés IA : {[kw[0] for kw in row['Mots_Clés']]}<br>"
        f"📌 Cluster géographique : {row['Cluster']}<br>"
        f"🧭 Latitude : {row['Latitude_Décimal']}<br>"
        f"🧭 Longitude : {row['Longitude_Décimal']}"
    )
    folium.Marker(
        location=[row['Latitude_Décimal'], row['Longitude_Décimal']],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color='blue' if row['Cluster'] == 0 else 'green' if row['Cluster'] == 1 else 'red')
    ).add_to(cluster_layer)

# 📌 Étape 6 : Tracé automatique de l’itinéraire du Nautilus
coordinates = list(zip(df['Latitude_Décimal'], df['Longitude_Décimal']))
folium.PolyLine(locations=coordinates, color='blue', weight=3, opacity=0.7).add_to(m)

# 📌 Étape 7 : Sauvegarde de la carte
df.to_csv('escales_analysees.csv', index=False)
m.save('carte_nautilus_ia.html')
print("✅ Carte générée avec IA dans 'carte_nautilus_ia.html' et analyse exportée en CSV")
```
